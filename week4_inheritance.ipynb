### Creating a derived class in C++

- **"is-a"** relationship
- auto gain access to **public & protected** members
- 
- the derived class purpose is to add **more funcs**
--- really help with code **reuse**

```python

#include <iostream>
#include <string>

class Automobile {
protected:  // unlike private --- the inheritance can access proteced part
    std::string brand_;
    int maxSpeed_;

public:
    Automobile(const std::string& b, int speed)
        : brand_(b), maxSpeed_(speed) {}

    void displayInfo() const {
        std::cout << "Brand: " << brand_
                  << ", Max Speed: " << maxSpeed_ << " km/h\n";
    }
};


class Car : public Automobile {
private:  // can NOT be access
    int numberOfDoors_;

public:
    Car(const std::string& b, int speed, int doors)
        : Automobile(b, speed), numberOfDoors_(doors) {}

    void displayInfo() const { // try Shadowing
        displayInfo();
        std::cout << "Doors: " << numberOfDoors_ << std::endl;
    }
};

int main() {
    Car myCar("Toyota", 180, 4);
    myCar.displayCarInfo();

    return 0;
}
```

### Shadowing

- use to hide the old function -- update it with newer ver in the child class
- can still be access --- car.automobile::display()  --- going back ::func()
